```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
df['Gender'] = np.where(df['Gender']=='Male',1,0)
```

```python
sns.scatterplot(df,x = 'Spending Score (1-100)',y='Age')
```

```python
sns.scatterplot(df,x = 'Spending Score (1-100)',y='Annual Income (k$)',hue = 'Gender')
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

X_train,X_test = train_test_split(df)
```

```python
cluster = KMeans(n_clusters=5)
```

```python
cluster.fit_transform(X_train)
```

```python
pred = cluster.predict(X_test)
```

```python
sns.scatterplot(X_test,x = 'Spending Score (1-100)',y='Annual Income (k$)',hue = pred)
```

```python
sns.scatterplot(X_test,x = 'Gender',y='Annual Income (k$)',hue = pred)
```
